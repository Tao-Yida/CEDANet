```python
# save as: visualize_numpy_labels.ipynb

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import cv2, os
```

```python
# ==== 配置路径 ====
img_dir = "../smoke-segmentation.v5i.coco-segmentation/test/"
mask_dir = "../smoke-segmentation.v5i.coco-segmentation/masks/"
base_name = "hoogovens_6_7__2024-11-03T07-38-21Z_frame_927_jpg.rf.b585fee20b605ecd1bdeb0553a2f4040"

img_path = os.path.join(img_dir, base_name + ".jpg")
mask_path = os.path.join(mask_dir, base_name + ".npy")

# ==== 加载图像和掩膜 ====
img = cv2.imread(img_path)[..., ::-1]
mask = np.load(mask_path)

# ==== 可视化 ====
cmap = ListedColormap(["black", "red", "lime"])

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(mask, cmap=cmap, vmin=0, vmax=2)
plt.title("Category Masks\n(0=Background, 1=High-opacity smoke, 2=Low-opacity smoke)")
# Legacy colorbar
plt.colorbar(ticks=[0, 1, 2], label="Category")
plt.clim(-0.5, 2.5)  # Set color limits for better visualization
plt.tight_layout()

plt.show()
```
